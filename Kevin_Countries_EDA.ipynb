## Looking differences depending on the country

### Drugs use vs. non drugs use

```python
country = ['USA' if c < -0.5 else 
           'New Zealand' if c > -0.5 and c < -0.4 else 
           'Other' if c > -0.4 and c < -0.2 else 
           'Australia' if c > -0.2 and c < 0 else 
           'Ireland' if c > 0 and c < 0.23 else 
           'Canada' if c > 0.23 and c < 0.9 else 
           'UK' 
           for c in drugs['Country']]
```

```python
drugs['Country'] = country
```

```python
drugs.Country.value_counts()
```

```python
uk_drugs = drugs.query("Country == 'UK'")
usa_drugs = drugs.query("Country == 'USA'")
other_drugs = drugs.query("Country == 'Other'")
canada_drugs = drugs.query("Country == 'Canada'")
aussie_drugs = drugs.query("Country == 'Australia'")
ireland_drugs = drugs.query("Country == 'Ireland'")
nz_drugs = drugs.query("Country == 'New Zealand'")
```

```python
plt.pie(uk_drugs.illegal_drugs.value_counts(),labels=["1", "0"], autopct='%1.1f%%')
plt.axis('equal')
plt.title("Balance of illegal drug use in the UK")
plt.show()
```

```python
plt.pie(usa_drugs.illegal_drugs.value_counts(),labels=["1", "0"], autopct='%1.1f%%')
plt.axis('equal')
plt.title("Balance of illegal drug use in the USA")
plt.show()
```

```python
plt.pie(other_drugs.illegal_drugs.value_counts(),labels=["1", "0"], autopct='%1.1f%%')
plt.axis('equal')
plt.title("Balance of illegal drug use in other countries")
plt.show()
```

```python
plt.pie(aussie_drugs.illegal_drugs.value_counts(),labels=["1", "0"], autopct='%1.1f%%')
plt.axis('equal')
plt.title("Balance of illegal drug use in Australia")
plt.show()
```
